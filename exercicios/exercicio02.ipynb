```python
import pandas as pd
```

```python
df = pd.read_csv('../dados/titanic.csv')
df
```

```python
''' Exercício 1 
# Quantas pessoas sobreviveram ao acidente?
'''

df.dropna()

df['sobreviveu']  = df['sobreviveu'].astype('category')
sobreviventes = df['sobreviveu'].value_counts()['sim']
print('{} pessoas sobreviveram'.format(sobreviventes))
```

```python
''' 
7) Supondo que o valor da tarifa no dataset está em libras, e que 1£ = R$ 6.43, 
crie uma nova coluna chamada **tarifa_real** que contenha a tarifa de embarque em reais.

'''
df2 = df.copy()
df2['tarifa_real'] = df2['tarifa']*6.43
df2.head()
```

```python
''' 
8) Agrupe pela coluna sobreviveu e calcule o número de passageiros por grupo (sim/nao)
'''



df['sobreviveu']  = df['sobreviveu'].astype('category')
df['sobreviveu'].value_counts()
```

```python
''' 
10) Agrupe pela classe e conte a quantidade de pessoas por classe
'''

df['classe']  = df['classe'].astype('category')
df[['classe','nome']].groupby('classe').agg(['count']) 
```

```python
''' 
11) Agrupe pela classe e pela coluna sobreviveu e 
conte a quantidade de pessoas que sobreviveu de acordo com cada classe. 
Quem sobreviveu mais? A classe mais alta ou a mais baixa?
'''

df['classe']  = df['classe'].astype('category')
df['sobreviveu']  = df['sobreviveu'].astype('category')

# se sobreviveu, valor é igual a 1, se morreu, valor igual a 0
def classifica_sobreviveu(row_value):
    if row_value == "sim":
        ticket = 1
    else:
        ticket = 0
    return ticket


# aplicamos a regra na coluna que nos interessa
df2['classifica_sobreviveu'] = df['sobreviveu'].apply(classifica_sobreviveu)
df2['classifica_sobreviveu']  = df2['classifica_sobreviveu'].astype('float64')


# agrupando por classe e sobreviveu
sobreviveu_classe = df2.groupby(["classe", "classifica_sobreviveu"]).agg( { 'classe': 'count', 'classifica_sobreviveu': 'sum'})

# transformando em percentual
sobreviveu_classe_pcts = sobreviveu_classe.assign(Percentage = lambda x: (x['classifica_sobreviveu'] / x['classifica_sobreviveu'].sum()))

print(sobreviveu_classe_pcts)

# criando lista que salva a classe e o percentual de sobreviventes em relação ao total de sobreviventes
tag_classe = 0
percent_sobrevive = []

for i in range(len(sobreviveu_classe_pcts)):
    if sobreviveu_classe_pcts.iloc[i, 1] > 0:
        tag_classe = tag_classe + 1
        percent_sobrevive.append((tag_classe, sobreviveu_classe_pcts.iloc[i, 2]))

# encontrando o maior valor da lista
classe_e_percentual_sobrevive = max(percent_sobrevive, key=lambda x: x[1])
index_classe_mais_sobrevive = percent_sobrevive.index(classe_e_percentual_sobrevive)

classe_mais_sobrevive = classe_e_percentual_sobrevive[0]
percentual_mais_sobrevive = "{:.2%}".format(classe_e_percentual_sobrevive[1])


print('{} das pessoas que sobreviveram estavam na {}a classe, e esta foi a classe em que mais sobreviveram pessoas'.format(percentual_mais_sobrevive, classe_mais_sobrevive))
```

```python
''' 
12) Agrupe pelo sexo e pela coluna sobreviveu e conte a quantidade de pessoas 
que sobreviveram de acordo com o sexo. Quem sobreviveu mais? Homens ou mulheres?
'''

#df = pd.DataFrame(np.array([[1,2,3,4],[4,5,6,1],['A','B','C','A'],['a','b','a','b']]).T,columns=['d1','d2','type','subtype'])
df['sexo']  = df['sexo'].astype('category')
df['sobreviveu']  = df['sobreviveu'].astype('category')
df.index
sobreviveu_sexo = df.groupby(["sexo", "sobreviveu"]).agg( { 'sexo': 'count', 'sobreviveu': 'count'})
print(sobreviveu_sexo)


mulher_sobreviveu = sobreviveu_sexo.loc['feminino', 'sim'].loc['sobreviveu']
homem_sobreviveu = sobreviveu_sexo.loc['masculino', 'sim'].loc['sobreviveu']

if mulher_sobreviveu > homem_sobreviveu:
    print('As mulheres foram as que mais sobreviveram ao desastre, com {} mulheres sobrevivendo'.format(mulher_sobreviveu))
else:
    print('Os homens foram os que mais sobreviveram ao desastre, com {} homens sobrevivendo'.format(homem_sobreviveu))

```
